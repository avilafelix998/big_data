```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("./data/Valve_Player_Data.csv")
df
```

```python
df_copia = df
df_copia
```

```python
df_copia.fillna(0)
```

```python
for col in df_copia.columns:
    print(f"'{col}'")
```

```python
df_copia.columns = df_copia.columns.str.strip()
```

```python
df_copia.drop(['Percent_Gain', 'Month_Year', 'URL'], axis=1, inplace=True)
```

```python
df_copia
```

```python
df_copia.to_csv('resultado.csv', index=False, sep=';')
```

```python
df_jugadores_anios = df_copia.copy()
```

```python
df_jugadores_anios.drop(['Gain', 'Peak_Players'], axis=1, inplace=True)
df_jugadores_anios = df_jugadores_anios[df_jugadores_anios["Game_Name"].str.contains("Counter Strike: Global Offensive", case=False, na=False)]
```

```python
df_jugadores_anios.to_csv('jugadores_anios.csv', index=False, sep=';')
```

```python
df_jugadores_mes = df_copia.copy()
df_jugadores_mes.drop(['Gain', 'Avg_players'], axis=1, inplace=True)
df_jugadores_mes = df_jugadores_mes[df_jugadores_mes["Game_Name"].str.contains("Counter Strike: Global Offensive", case=False, na=False)]
df_jugadores_mes.to_csv('jugadores_mes.csv', index=False, sep=';')
```

```python
df_gain = df_copia.copy()
df_gain.drop(['Peak_Players', 'Avg_players'], axis=1, inplace=True)
df_gain = df_gain[df_gain["Game_Name"].str.contains("Counter Strike: Global Offensive", case=False, na=False)]
df_gain.to_csv('gain.csv', index=False, sep=';')
```

```python
df_top_promedio = df_copia.copy()
df_top_promedio = df_top_promedio[df_top_promedio["Game_Name"].str.contains("Counter Strike: Global Offensive", case=False, na=False)]
df_top_promedio.drop(['Gain', 'Peak_Players','Date'], axis=1, inplace=True)
df_top_promedio.to_csv('top_promedio.csv', index=False, sep=';')
```

```python

df['Date'] = pd.to_datetime(df['Date'])

df_csgo = df[df['Game_Name'] == "Counter Strike: Global Offensive"]
df_csgo = df_csgo.sort_values(by='Date')

plt.figure(figsize=(10, 5))
plt.plot(df_csgo['Date'], df_csgo['Avg_players'], marker='o', color='steelblue')
plt.title('Promedio de Jugadores por Año – CS:GO')
plt.xlabel('Fecha')
plt.ylabel('Jugadores Promedio')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.plot(df_csgo['Date'], df_csgo['Peak_Players'], marker='x', color='darkorange')
plt.title('Pico de Jugadores por Mes – CS:GO')
plt.xlabel('Fecha')
plt.ylabel('Pico de Jugadores')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
fechas = df_csgo['Date'].dt.strftime('%Y-%m')

plt.figure(figsize=(12, 6))
plt.bar(fechas, df_csgo['Gain'], color='skyblue')
plt.title('Ganancia o Pérdida de Jugadores en CS:GO')
plt.xlabel('Fecha')
plt.ylabel('Gain (jugadores)')

n = 6  
plt.xticks(ticks=range(0, len(fechas), n), labels=fechas[::n], rotation=45)

plt.tight_layout()
plt.grid(axis='y')
plt.show()
```

```python
df_avg_by_game = df.groupby('Game_Name')['Avg_players'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(8, 8))
plt.pie(df_avg_by_game, labels=df_avg_by_game.index, autopct='%1.1f%%', startangle=140)
plt.title('Top 10 Juegos por Promedio de Jugadores')
plt.axis('equal')
plt.show()
```
