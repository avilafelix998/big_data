```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("./data/Valve_Player_Data.csv")
df
```

```python
df_copia = df
df_copia
```

```python
df_copia.fillna(0)
```

```python
df_copia.drop(['Percent_Gain', 'Month_Year', 'URL'], axis=1, inplace=True)
```

```python
df_jugadores_anios.to_csv('jugadores_anios.csv', index=False, sep=',')
```

```python
df_jugadores_mes = df_copia.copy()
df_jugadores_mes.drop(['Gain', 'Avg_players'], axis=1, inplace=True)
df_jugadores_mes = df_jugadores_mes[df_jugadores_mes["Game_Name"].str.contains("Counter Strike: Global Offensive", case=False, na=False)]
df_jugadores_mes.to_csv('jugadores_mes.csv', index=False, sep=',')
```

```python
df_gain = df_copia.copy()
df_gain.drop(['Peak_Players', 'Avg_players'], axis=1, inplace=True)
df_gain = df_gain[df_gain["Game_Name"].str.contains("Counter Strike: Global Offensive", case=False, na=False)]
df_gain.to_csv('gain.csv', index=False, sep=',')
```

```python
df_top_promedio = df_copia.copy()
df_top_promedio.drop(['Gain', 'Peak_Players', 'Date'], axis=1, inplace=True)
df_top_promedio.to_csv('top_promedio.csv', index=False, sep=',')
```

```python

df['Date'] = pd.to_datetime(df['Date'])

df_csgo = df[df['Game_Name'] == "Counter Strike: Global Offensive"]
df_csgo = df_csgo.sort_values(by='Date')

plt.figure(figsize=(10, 5))

# Gráfico de líneas
plt.plot(df_csgo['Date'], df_csgo['Avg_players'], marker='o', color='steelblue')

# Título y etiquetas con color personalizado
plt.title('Promedio de Jugadores por Año – CS:GO', color='#c7d5e0')
plt.xlabel('Fecha', color='#c7d5e0')
plt.ylabel('Jugadores Promedio', color='#c7d5e0')

# Ejes con color
plt.xticks(rotation=45, color='#c7d5e0')
plt.yticks(color='#c7d5e0')

# Fondo del gráfico y figura
plt.gca().set_facecolor('#171a21')     # fondo del gráfico
plt.gcf().set_facecolor('#171a21')     # fondo general

# Grilla
plt.grid(True, color='#444')

plt.tight_layout()
plt.show()

```

```python
plt.figure(figsize=(10, 5))

# Línea del gráfico
plt.plot(df_csgo['Date'], df_csgo['Peak_Players'], marker='x', color='darkorange')

# Título y etiquetas en color personalizado
plt.title('Pico de Jugadores por Mes – CS:GO', color='#c7d5e0')
plt.xlabel('Fecha', color='#c7d5e0')
plt.ylabel('Pico de Jugadores', color='#c7d5e0')

# Ejes (números del eje x e y) en color personalizado
plt.xticks(rotation=45, color='#c7d5e0')
plt.yticks(color='#c7d5e0')

# Fondo del gráfico y de la figura
plt.gca().set_facecolor('#171a21')     # fondo del área del gráfico
plt.gcf().set_facecolor('#171a21')     # fondo general

# Grilla con color tenue
plt.grid(True, color='#444')

plt.tight_layout()
plt.show()

```

```python


fechas = df_csgo['Date'].dt.strftime('%Y-%m')

fig, ax = plt.subplots(figsize=(12, 6))

ax.bar(fechas, df_csgo['Gain'], color='skyblue')

# Título y etiquetas con color personalizado
ax.set_title('Ganancia o Pérdida de Jugadores en CS:GO', color='#c7d5e0')
ax.set_xlabel('Fecha', color='#c7d5e0')
ax.set_ylabel('Gain (jugadores)', color='#c7d5e0')

# Ejes en color personalizado
ax.tick_params(axis='x', colors='#c7d5e0', rotation=45)
ax.tick_params(axis='y', colors='#c7d5e0')

# Marcas de tiempo en eje X
n = 6
ax.set_xticks(range(0, len(fechas), n))
ax.set_xticklabels(fechas[::n])

# Fondo y grilla
fig.patch.set_facecolor('#171a21')  # fondo general
ax.set_facecolor('#171a21')         # fondo del área del gráfico
ax.grid(axis='y', color='#444')     # grilla sutil

plt.tight_layout()
plt.show()

```

```python

df_avg_by_game = df.groupby('Game_Name')['Avg_players'].mean().sort_values(ascending=False).head(10)

fig, ax = plt.subplots(figsize=(8, 9))
wedges, texts, autotexts = ax.pie(
    df_avg_by_game,
    labels=df_avg_by_game.index,
    autopct='%1.1f%%',
    startangle=140,
    textprops=dict(color='#c7d5e0')  # color del texto de etiquetas y porcentajes
)

# Título con color personalizado
plt.title('Top 10 Juegos por Promedio de Jugadores', color='#c7d5e0')

# Fondo
fig.set_facecolor('#171a21')  # fondo oscuro estilo Steam
plt.axis('equal')  # mantener proporciones
plt.show()

```
